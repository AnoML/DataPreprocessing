```python
from pandas import DataFrame
from pandas import read_csv
from pandas import set_option
```

```python
df = read_csv('anoml_dataset.csv', header=0)
```

## Remove Presentage marks, commas and converting to float

```python
half_rows = df.shape[0]//2
for i in range(df.shape[1]):
    selected_col = df.iloc[:, i]
    col_type = type(selected_col[0])
    
    if(col_type == str):
        if(selected_col.iloc[0].endswith('%')):
            df.iloc[:, i] = df.iloc[:, i].str.rstrip('%').astype('float') / 100.0
       
        elif(selected_col.iloc[0].find(",") != -1 or selected_col.iloc[-1].find(",") != -1 or selected_col.iloc[100].find(",") != -1):
            df.iloc[:, i] = df.iloc[:, i].str.replace(',', '').astype('float')
        
        elif(selected_col.iloc[0].find("B") != -1 or selected_col.iloc[-1].find("B") != -1 or selected_col.iloc[half_rows].find("MB") != -1):
            for index, row in DataFrame(df.iloc[:, i]).iterrows():
                if (row.values[0].find("MB") != -1):
                    df.iloc[[index],[i]] = float(row.values[0].replace('MB','').replace(',',''))*1000000 
                    
                elif (row.values[0].find("KB") != -1):
                    df.iloc[[index],[i]] = float(row.values[0].replace('KB','').replace(',',''))*1000 
                
                elif (row.values[0].find("B") != -1):
                    df.iloc[[index],[i]] = float(row.values[0].replace('B','').replace(',','')) 
                    
```
